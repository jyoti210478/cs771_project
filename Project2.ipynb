```python
import numpy as np
import pandas as pd
import time as tm
import warnings
warnings.filterwarnings('ignore')
```

```python
with open('/content/dict', 'r') as file:
    words = [line.strip() for line in file]
```

```python
data = pd.DataFrame({'respective_word': words})
```

```python
data.head()
```

```python
def generate_sorted_unique_char_bigrams(word):
        char_bigrams = [word[i:i+2] for i in range(len(word)-1)]
        unique_bigrams = set(char_bigrams)
        unique_bigrams=sorted(unique_bigrams)
        return unique_bigrams
```

```python
data['bigram_list'] = data['respective_word'].apply(generate_sorted_unique_char_bigrams)
# Convert the bigram lists to a single string for display purposes (optional)
#data['bigram_list'] = data['bigram_list'].apply(lambda x: ','.join(x))
# This is the DataFrame with bigram_list and respective_word columns
# Function to truncate bigram lists to a maximum length of 5
def truncate_bigram_list(bigram_list):
    t=min(len(bigram_list), 5)
    return bigram_list[:t]

# Apply the function to the bigram_list column
data['bigram_list'] = data['bigram_list'].apply(truncate_bigram_list)
```

```python
all_bigrams = sorted(set().union(*data['bigram_list']))

    # Create bigram vectors
for bigram in all_bigrams:
    data[bigram] = data['bigram_list'].apply(lambda x: 1 if bigram in x else 0)
```
